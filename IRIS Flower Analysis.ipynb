```python
#Scatter plot
colors =['red','blue','yellow']   #manually assign the color
Species=['Iris-setosa','Iris-virginica','Iris-versicolor']
```

```python
for i in range(3):
    x =iris[iris['Species']  == Species[i]]
    plt.scatter(x['SepalLengthCm'],x['SepalWidthCm'], c = colors[i], label =Species[i])
plt.xlabel("SepalLenth")
plt.ylabel("SepalWidth")
plt.legend()
sns.set_style("darkgrid")
plt.show()
```

```python
for i in range(3):
    x =iris[iris['Species']  == Species[i]]
    plt.scatter(x['PetalLengthCm'],x['PetalWidthCm'], c = colors[i], label =Species[i])
plt.xlabel("PetalLenth")
plt.ylabel("PetalWidth")
plt.legend()
sns.set_style("darkgrid")
plt.show()
```

```python
#trying to 3d Scatter
sns.set_style("darkgrid");
sns.FacetGrid(iris, hue='Species',height=3) \
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm", "PetalLengthCm") \
    .add_legend()
plt.show()
```

```python
                                  # pair plot
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue = 'Species', height =3)
plt.grid()
plt.show()
```

```python
#what about 1 d scatter by using 1 feature
#1 d scatter by using petal length
import numpy as np
iris_setosa = iris.loc[iris["Species"] == "Iris-setosa"]
iris_virginica = iris.loc[iris["Species"] == "Iris-virginica"]
print(iris_virginica)
iris_versicolor = iris.loc[iris["Species"] == "Iris-versicolor"]
# plotting
plt.plot(iris_setosa["PetalLengthCm"])
plt.plot(iris_virginica["PetalLengthCm"])
plt.plot(iris_versicolor["PetalLengthCm"])
plt.show()
print("mean")
print(iris_setosa["PetalLengthCm"].mean())
```

```python
sns.FacetGrid(iris , hue ="Species" , height = 8) \
   .map(sns.distplot, "PetalLengthCm") \
   .add_legend();
plt.show();
```

```python
print("mean")
print(np.mean(iris_setosa["PetalLengthCm"]))
print(np.mean(iris_virginica["PetalLengthCm"]))
print(np.mean(iris_versicolor["PetalLengthCm"]))
```

```python
#outlier may error comes by import 51th element may be also by error

print(np.mean(np.append(iris_setosa["PetalLengthCm"],50)))   # and bcz of one error mean jumps highly
```

```python
print("Standard deviation")
print(np.std(iris_setosa["PetalLengthCm"]))
print(np.std(iris_virginica["PetalLengthCm"]))
print(np.std(iris_versicolor["PetalLengthCm"]))
```

```python
#median
print("median")
print(np.median(iris_setosa["PetalLengthCm"]))
print(np.median(iris_virginica["PetalLengthCm"]))
print(np.median(iris_versicolor["PetalLengthCm"])) 
    
```

```python
print(np.median(np.append(iris_setosa["PetalLengthCm"],50)))
print(np.median(np.append(iris_virginica["PetalLengthCm"],50)))
print(np.median(np.append(iris_versicolor["PetalLengthCm"],50)))   

# its much better than mean std and varience   same behavious after addend or error
```

```python
#what happen if 50% of data currepted or more then median become also be worst
#90th percentile

print("/npercentile")
print(np.percentile(iris_setosa["PetalLengthCm"],90))
print(np.percentile(iris_virginica["PetalLengthCm"],90))
print(np.percentile(iris_versicolor["PetalLengthCm"],90)) 
```

```python
print("90th/Quantiles")
print(np.percentile(iris_setosa["PetalLengthCm"], np.arange(0,100,25)))  #pericentile between 0 to 100 of gap 25
print(np.percentile(iris_virginica["PetalLengthCm"], np.arange(0,100,25)))
print(np.percentile(iris_versicolor["PetalLengthCm"], np.arange(0,100,25)))
```

```python
from statsmodels import robust
print("/n median absolute deviation ")
print(robust.mad(iris_setosa["PetalLengthCm"]))
print(robust.mad(iris_virginica["PetalLengthCm"]))
print(robust.mad(iris_versicolor["PetalLengthCm"])) 
```
